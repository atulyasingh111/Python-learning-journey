```python
#3. String Boolean Methods
```

```python
string= "Atulya Singh123"
```

```python
#type string. and press tab to see the available functions
```

```python
print(string.isalpha())       #contains alphabet
print(string.isalnum())       #contains alphabet and numbers
print(string.islower())       #lower case
print(string.isupper())       #upper case
print(string.istitle())       #first letter is capital
print(string.isspace())       #contains spaces
print(string.startswith("A")) #it is case sensitive
print(string.endswith("H"))   #it is also case sensitive
```
