```python
#2. DICTIONARIES: is a collection which is unordered, changeable and indexed, stored in curly brackets with key and value pairs
```

```python
dict()      #it is a built in function 
type(dict())
```

```python
dict_ex= {}
type(dict_ex)
```

```python
set_ex= {1,2,3,4}
type(set_ex)
```

```python
dict1= {1:"A", 2: "B", 3: "C", 4: "D"}
type(dict1)
```

```python
#indexing is done using keys
print(dict1[1])
```

```python
for x in dict1:      #by default the key values are printed
    print(x)
```

```python
#to print only values
for x in dict1.values():
    print(x)
```

```python
#to print both key and values
for x in dict1.items():
    print(x)
```

```python
dict1[5]="E"
```

```python
print(dict1)
```

```python
#NESTED DICTIONARIES
```

```python
car_model1= {'Mercedez': 1960}
car_model2= {'Toyota': 1971}
car_model3= {'Creta': 1950}

car_dict= {'car1': car_model1, 'car2':car_model2, 'car3':car_model3 }
print(car_dict)
```

```python
#Acessing the items in the dictionaries
print(car_dict['car1'])
```

```python
print(car_dict['car1']['Mercedez'])
```
